# 行列演算の基礎

## 

- ベクトルの定義
- 行列の定義
- 転置
- 逆行列
- 行列積

```python
import numpy as np
```

```python
# ベクトルの定義
x = np.array([[1], [2], [3], [4]])
print(x)
```

```python
# 行列の定義
X = np.array([[1,2],[3,4]])
print(X)
```

```python
# 転置
Xt = X.T
print(Xt)
```

```python
# 逆行列
# linear algebra: 線形代数
X_inv = np.linalg.inv(X)
print(X_inv)
```

```python
# 行列積
XX_inv = np.dot(X, X_inv)
print(XX_inv)
```

#### 行列Xの1列目1はダミーの変数より、変数は2つ

```python
# Xの定義
X = np.array([
    [1,2,3],
    [1,2,5],
    [1,3,4],
    [1,5,9]
])
```

```python
print(X)
```

```python
# yの定義
y = np.array([
    [1],
    [5],
    [6],
    [8]
])
```

```python
print(y)
```

```python
# Step1
XtX = np.dot(X.T, X)
```

```python
print(XtX)
```

```python
# Step2
XtX_inv = np.linalg.inv(XtX)
```

```python
print(XtX_inv)
```

```python
# Step3
Xty = np.dot(X.T, y)
```

```python
print(Xty)
```

```python
w = np.dot(XtX_inv, Xty)
```

```python
print(w)
```

## Scikit-learnで実装

```python
import sklearn
```

```python
# 重回帰分析のみ読み込み
from sklearn.linear_model import LinearRegression
```

```python
# モデルの宣言
# model = LinearRegression()
model = LinearRegression(fit_intercept=False)
```

```python
# モデルの学習 ← パラメータの調整
model.fit(X,y)
```

```python
# 調整後のパラメータ
model.coef_
```

```python
model.intercept_
```

```python
# 予測精度 ← 決定係数
model.score(X, y)
```

```python
# 予測値の計算
x = np.array([[1,2,3]])
```

```python
y_pred = model.predict(x)
y_pred
```
