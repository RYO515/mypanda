```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(font=["Meiryo", "Gothic", "Hiragino Maru Gothic Pro"])

data = {
    "A案" : [1,10,1,10,1,10,1,10],
    "B案" : [5,5,5,5,6,6,6,6],
    "C案" : [1,2,3,4,7,8,9,10]
}

df = pd.DataFrame(data)
df
```

```python
# データフレーム.plot.hist(bins=階級を区切る範囲)
bins=[1,3,5,7,9,11]

df.plot.hist(bins=bins)
# titleは「何のデータのためにグラフを作ったのか」を表すので重要
plt.title("ケーキの感想はどのように違うか？")
plt.show()
```

```python
df["A案"].plot.hist(bins=bins)
plt.title("A案のケーキの感想")
plt.show()

df["B案"].plot.hist(bins=bins)
plt.title("B案のケーキの感想")
plt.show()

df["C案"].plot.hist(bins=bins)
plt.title("C案のケーキの感想")
plt.show()
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font=["Meiryo", "Gothic", "Hiragino Maru Gothic Pro"])

data = {
    "身長" :  [163.6, 172.6, 163.7, 169.9, 173.9, 166.2, 176.7, 165.4],
    "体重" :  [50.5, 63.3, 48.5, 59.8, 69.8, 53.7, 70.3, 51.2]
}

df = pd.DataFrame(data)
df
```

```python
# 散布図: 2種類のデータの関係性がわかるグラフ
# データフレーム.plot.scatter(x="横軸の列名", y="縦軸の列名", c="色")
# black=k, red=r, blue=b, green=g, yellow=y, cyan=c, magenta=m
df.plot.scatter(x="身長", y="体重", c="b")
plt.title("身長と体重の関係")
plt.show()
```

```python
# マーカー
# plt.plot(x座標, y座標, c="色", marker="マーカー", markersize="マーカーサイズ")
# 3行目のX座標はdf.iloc[3]["身長"]、3行目のY座標はdf.iloc[3]["体重"]
# sns.set_context("グラフの文字サイズ")
# paper < notebook < talk < poster 文字サイズの順
sns.set_context("talk")
df.plot.scatter(x="身長", y="体重", c="b", figsize=(12,8))

x=df.iloc[3]["身長"]
y=df.iloc[3]["体重"]
plt.plot(x, y, c="b", marker="X", markersize=15)

plt.title("私はどこにいるか")
plt.show()
```

```python
df.plot.scatter(x="身長", y="体重", c="b", figsize=(12,8))
plt.title("私はどこにいるか？")

X=df.iloc[3]["身長"]
Y=df.iloc[3]["体重"]
plt.plot(x, y, c="r", marker="X", markersize=15)

# グラフに線を引く
plt.axvline(x=X, c="r", linestyle="--")
plt.axhline(y=Y, c="r", linestyle="--")

plt.show()
```
